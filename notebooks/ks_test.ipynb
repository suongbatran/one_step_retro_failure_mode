```python
import pandas as pd
import json
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
```

```python
rxn_complex = pd.read_csv("../data_uspto/complexity_features/rxn_complexity.csv")
train  = pd.read_csv("../data_uspto/75_5_20_same_complex_doc_split/raw_train.csv")
test = pd.read_csv("../data_uspto/75_5_20_same_complex_doc_split/raw_test.csv")
```

```python
train_w_complex = train.merge(rxn_complex, on=["id", "rxn_smiles", "rxn_smiles_rxnmapper"], how="left")
test_w_complex = test.merge(rxn_complex, on=["id", "rxn_smiles", "rxn_smiles_rxnmapper"], how="left")
```

```python
cols = ['num_react_atom_forward','num_react_atom_forward_rxnmapper',
       'num_react_atom_retro', 'num_react_atom_retro_rxnmapper',
       'num_change_ring', 'num_change_ring_rxnmapper']

summary_df = pd.DataFrame()

for j, col in enumerate(cols):
    data1 = train_w_complex[col].dropna().values
    data2 = test_w_complex[col].dropna().values
    res = ks_2samp(data1, data2, alternative="two-sided")

    idx = len(col)+j
    summary_df.loc[idx, "complexity_metric"] = col
    summary_df.loc[idx, "ks_statistic"] = res.statistic

summary_df
```
